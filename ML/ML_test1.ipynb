В этом задании вам необходимо реализовать алгоритм МНК (метод наименьших квадратов) в виде функции с использованием библиотеки numpy.

Входные данные:

- x_: временной ряд x длиной 30 точек, сгенерированный с помощью функции linspace;

- y_: временной ряд y длиной 30 точек, сгенерированный с помощью функции linspace с коэффициентами a и b.

```python
import numpy as np
```

```python
def mnk(x_,y_):
    xi = sum(x_)
    yi = sum(y_)
    xi2 = sum(x_**2)
    xi_yi = sum(x_*y_)
    n = x_.size
    A = np.array([[xi2, xi], [xi, n]])
    B = np.array([xi_yi, yi])
    X = np.linalg.solve(A, B)
    a = X[0]
    b = X[1]
    return a,b
```

```python
x = np.linspace(0, 10, 30)
y = np.linspace(0, 10, 30)*(10+4*np.random.rand(30)) + (10+4*np.random.rand(30))

a,b = mnk(x,y)

print(a, b)
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
print(mnk(x, y))
```
