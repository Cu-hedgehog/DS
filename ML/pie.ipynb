```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data/Electric_Car.csv')
df.info()
```

```python
df1 = df['Brand'].value_counts()
rest = df1.iloc[5:].sum()
df1 = df1.iloc[:5]
df1['Rest'] = rest
#df.sort_values(by='TopSpeed_KmH')
df1
```

```python
fig = plt.figure(figsize=(5, 5))
axes = fig.add_axes([0, 0, 1, 1])
axes.pie(
    df1,
    labels=df1.index,
    autopct='%.1f%%',
);
```
